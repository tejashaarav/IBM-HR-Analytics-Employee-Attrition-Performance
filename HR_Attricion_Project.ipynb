```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
columnle = ['Attrition','BusinessTravel', 'Department','EducationField','Gender','JobRole','MaritalStatus','Over18','OverTime']

for i in columnle:
    df[i] = le.fit_transform(df[i])


df
```

```python
from scipy.stats import zscore
z = np.abs(zscore(df))
z
```

```python
x = df.drop('Attrition',axis =1)
y = df['Attrition']
x
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
```

```python
x = ss.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,random_state = 42,test_size = 0.2)
xtrain
```

```python
ytest.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
#logistic regression 
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

lr = LogisticRegression()
lr = lr.fit(xtrain,ytrain)
ytest_lr = lr.predict(xtest)
print("accuracy:",accuracy_score(ytest,ytest_lr) )

print("Classification Report:",classification_report(ytest_lr,ytest) )
```

```python
#RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


rf = RandomForestClassifier()
rf = rf.fit(xtrain,ytrain)
ytest_rf = rf.predict(xtest)
print("accuracy:",accuracy_score(ytest,ytest_rf) )

print("Classification Report:",classification_report(ytest_rf,ytest) )
```

```python
#SVC
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

svc = SVC()
svc = svc.fit(xtrain,ytrain)
ytest_svc = svc.predict(xtest)
print("accuracy:",accuracy_score(ytest,ytest_svc) )

print("Classification Report:",classification_report(ytest_svc,ytest) )
```

```python
#DecisionTreeClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

dtc = DecisionTreeClassifier()
dtc = dtc.fit(xtrain,ytrain)
ytest_dtc = dtc.predict(xtest)
print("accuracy:",accuracy_score(ytest,ytest_dtc) )

print("Classification Report:",classification_report(ytest_dtc,ytest) )
```

```python
#KNeighborsClassifier

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

knc = KNeighborsClassifier()
knc = knc.fit(xtrain,ytrain)
ytest_knc = knc.predict(xtest)
print("accuracy:",accuracy_score(ytest,ytest_knc) )

print("Classification Report:",classification_report(ytest_knc,ytest) )
```
